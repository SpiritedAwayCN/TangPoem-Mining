```python
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix, csr_matrix, csc_matrix
```

```python
table = pd.read_excel('data/poem_v2.xlsx')
word_list_df = pd.read_excel('data/wordlist_v2.xlsx')
word_list_df
```

```python
max_length = word_list_df.word.str.len().max()

def cut(string, word_list, max_length, regen=False):
    words = []
    index = 0
    max_l = len(string)
    while index < max_l:
        for i in range(max_length, 0, -1):
            if string[index: index + i] in word_list:
                s = string[index: index + i]
#                 if s != '一作':
                words.append(string[index: index + i])
                index += i
                break
        else:
            index += 1
    if regen and string[:2] == '##':
        words = words * 3
    return words

word_list = set(word_list_df.simple.values)
table['words'] = table['simple'].apply(lambda x: ' '.join(cut(x, word_list, max_length, True)))
table
```

```python
split_words = table[table['line_number']!=-1]['words'].str.split(' ', expand=True).stack().rename('word').reset_index()
# 以上剔除了作者行
split_words = split_words[split_words['word'] != '']
new_data = pd.merge(table['Poem_id'], split_words, left_index=True, right_on='level_0')
new_data
```

## 诗歌的tf-idf计算

```python
temp_df = new_data[['Poem_id', 'word']].drop_duplicates()
df_series = temp_df.groupby('word')['Poem_id'].apply(lambda x: x.shape[0]).rename('idf')
N = new_data['Poem_id'].value_counts().shape[0]
idf = pd.DataFrame(df_series)
idf['idf'] = np.log(N / idf)
idf
```

```python
word_dict = {}
word_dictr = []
for i, tup in enumerate(idf.itertuples()):
    word_dict[tup[0]] = i
    word_dictr.append(tup[0])
```

```python
tf_att = new_data.groupby(['Poem_id', 'word']).apply(lambda x: x.shape[0]).rename('tf')
tf_att
```

```python
idf = idf['idf'].values
```

```python
poemid_dict = {}
poemid_rdict = []
i = 0
for pid in table.Poem_id.values:
    if not pid in poemid_dict.keys():
        poemid_dict[pid] = i
        poemid_rdict.append(pid)
        i += 1
```

```python
# 仅诗词作为上下文时
d_index = []
t_index = []
tfidf_v = []

for (pid, word), value in tf_att.items():
    tid = word_dict[word]
    d_index.append(poemid_dict[pid])
    t_index.append(tid)
    tfidf_v.append(value * idf[tid])

poem_tfidf = coo_matrix((tfidf_v, (d_index, t_index)), shape=(len(poemid_dict), len(word_dict))).tocsr()
poem_tfidf
```


## 近义词TF-IDF计算

```python
table['words'] = table['simple'].apply(lambda x: ' '.join(cut(x, word_list, max_length)))

split_words = table[table['line_number']!=-1]['words'].str.split(' ', expand=True).stack().rename('word').reset_index()
split_words = split_words[split_words['word'] != '']
new_data = pd.merge(table['Poem_id'], split_words, left_index=True, right_on='level_0')

df = pd.DataFrame(new_data.groupby('word').apply(lambda x: x.shape[0])).reset_index()
sym_set = set(df[df[0] > 10].word.values)
sym_set.remove('一作')

table
```

```python
x_indexes = []
y_indexes = []

for row in table.itertuples():
    if row.line_number == -1:
        continue
    sentence = row.words.split(' ')
    n = len(sentence)
    for i in range(n):
        if not sentence[i] in sym_set:
            continue
        for j in range(n):
            if i == j:
                continue
            x_indexes.append(word_dict[sentence[i]])
            y_indexes.append(word_dict[sentence[j]])
N = len(word_dict)
            
vals = [1] * len(x_indexes)
sym_tf = coo_matrix((vals, (x_indexes, y_indexes)), shape=(N, N)).tocsr()
sym_tf
```

```python
sym_idf = np.zeros((N,))
for i in sym_tf.nonzero()[1]:
    sym_idf[i] += 1
sym_idf = np.log(len(sym_set) / (sym_idf + 1))
sym_idf
```

```python
from scipy import sparse
sym_tfidf = sym_tf * sparse.diags(sym_idf, shape=(N, N))
sym_tfidf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(sym_tfidf, dense_output=False)
sim
```

```python
def find_sum(word):
    candi = []
    for token in sym_set:
        for c1 in word:
            for c2 in token:
                if c1 == c2:
                    candi.append(token)
                    break
        else:
            continue
        break
    pred = {}
    wid = word_dict[word]
    for w in candi:
        pred[w] = sim[wid, word_dict[w]]
    return pd.Series(pred).rename('cosine')
```

```python
from sklearn.preprocessing import normalize

def judge_same_word(str1, str2):
    for c in str1:
        for cc in str2:
            if c == cc:
                return True
    return False

def query_final(key_words, hit=False, author=None):
    x_indexes = []
    vals = []
    for word in key_words:
        wid = word_dict[word]
        x_indexes.append(wid)
        vals.append(1)
        
        for sym_id in sim[wid, :].nonzero()[1]:
            sym_word = word_dictr[sym_id]

            if sym_word in key_words:
                continue

            val = sim[wid, sym_id]
            if judge_same_word(word, sym_word) and val > 0.08:
                print(sym_word)
                x_indexes.append(sym_id)
                vals.append(val)
            elif val > 0.3:
                print(sym_word)
                x_indexes.append(sym_id)
                vals.append(val * 0.1)
    weight_vec = coo_matrix((vals, (x_indexes, [0] * len(vals))), shape=(len(word_dictr), 1)).tocsr()
    res = poem_tfidf * weight_vec
    ret_num = res.data.shape[0]
    res = res.toarray().squeeze()
    ans = np.argsort(res)[::-1][:ret_num]
    if not author is None:
        ans = filter(lambda x: table[(table['Poem_id']==poemid_rdict[x]) & (table['line_number']==-1)].simple.values[0] == f'$${author}', ans)
        ans = np.array(list(ans))
    
    if hit == False:
        return zip(ans, res[ans])
    
    final_poems = set()
    for pid_r in ans:
        for pr in thresed[pid_r, :].nonzero()[1]:
            ppid = poemid_rdict[pr]
            if (not author is None) and table[(table['Poem_id']==ppid) & (table['line_number']==-1)].simple.values[0] != f'$${author}':
                continue
            final_poems.add(pr)

    final_poems = np.array(list(final_poems))
    temp = thresed[final_poems, :]
    X = temp[:, final_poems]
    
    temp = poem_sim[final_poems, :]
    XX = temp[:, final_poems]
    XX = normalize(XX).T #要转置，nomarlize是对行归一化（归一化前的XX是对称矩阵）
    
    #y/new_y 主特征向量（行向量）
    #X二值化的
    #XX列归一化后的相似度矩阵
    

    n = len(final_poems)
    y = np.full((1, n), 1)
    y = csr_matrix(y)

    last_y = csr_matrix(np.zeros((1, n)))
    for i in range(70):
        new_y = normalize(y)
        dis = np.sum((new_y - last_y).toarray()**2)
        print(i, dis)
        if  np.sqrt(dis) <= 1e-3:
            break
        last_y = new_y
        y = new_y * X
    print(new_y.shape, X.shape)
    
#     print(XX)
#     new_y = sparse.diags(new_y.toarray().squeeze(), shape=(n, n))
    new_y = new_y * XX
    print(new_y.shape)
    idx = []
    for pid_r in ans:
        for i, p in enumerate(final_poems):
            if p == pid_r:
                idx.append(i)
                break
    return zip(ans, zip(res[ans], new_y.toarray().squeeze()[idx]))
```
